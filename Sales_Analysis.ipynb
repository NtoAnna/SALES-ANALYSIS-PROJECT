```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/bright sales analysis/Sales Case 3 Study.csv')

df["Date"] = pd.to_datetime(df["Date"])
df.sort_values("Date", inplace=True)

plt.figure(figsize=(12, 6))
plt.plot(df["Date"], df["Sales"], marker="o", linestyle="-", label="Daily Sales")
plt.xlabel("Date")
plt.ylabel("Sales (Rand)")
plt.title("Daily Sales Trend")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
df["Gross_Profit_Percentage"] = ((df["Sales"] - df["Cost Of Sales"]) / df["Sales"]) * 100

plt.figure(figsize=(12, 6))
plt.plot(df["Date"], df["Gross_Profit_Percentage"], marker="o", linestyle="-", color="green")
plt.xlabel("Date")
plt.ylabel("Gross Profit %")
plt.title("Gross Profit Percentage Trend")
plt.xticks(rotation=45)
plt.show()
```

```python
df["Price_Per_Unit"] = df["Sales"] / df["Quantity Sold"]

plt.figure(figsize=(10,5))
plt.scatter(df["Price_Per_Unit"], df["Quantity Sold"], alpha=0.7, color="purple")
plt.xlabel("Price Per Unit")
plt.ylabel("Quantity Sold")
plt.title("Price Elasticity Scatter Plot")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/bright sales analysis/Sales Case 3 Study.csv')
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
df.sort_values("Date", inplace=True)

plt.figure(figsize=(12, 6))
plt.plot(df["Date"], df["Sales"], marker="o", linestyle="-", label="Daily Sales")
plt.xlabel("Date")
plt.ylabel("Sales (Rand)")
plt.title("Daily Sales Trend")
plt.xticks(rotation=45)

for i in df[df["Sales"] > df["Sales"].quantile(0.95)].index:
    plt.annotate(f"{df.loc[i, 'Sales']:.0f}",
                 (df.loc[i, "Date"], df.loc[i, "Sales"]),
                 textcoords="offset points",
                 xytext=(0,10),
                 ha="center")

plt.legend()
plt.show()
```
